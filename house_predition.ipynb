```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('train.csv')
train.head()
```

```python
train.LotFrontage.isna().sum()
```

```python
train=train.drop(["Id","Utilities","Street"],axis=1)
```

```python
train.shape
```

```python
train = train.fillna({
    'PoolQC': 'None',
    'MiscFeature': 'None',
    'Alley': 'None',
    'Fence': 'None',
    'FireplaceQu': 'None',
    'GarageType': 'None',
    'GarageFinish': 'None',
    'GarageQual': 'None',
    'GarageCond': 'None',
    'BsmtQual': 'None',
    'BsmtCond': 'None',
    'BsmtExposure': 'None',
    'BsmtFinType1': 'None',
    'BsmtFinType2': 'None',
    'MasVnrType': 'None'
})
```

```python
train.isna().sum()[train.isna().sum()>0]
```

```python
train['MasVnrArea']=train["MasVnrArea"].fillna(0.0)
```

```python
train
```

```python
train.GarageYrBlt.value_counts()
```

```python
train['GarageYrBlt'] = train['GarageYrBlt'].fillna(0)
```

```python
train.LotFrontage.value_counts()
```

```python
train["LotFrontage"]=train['LotFrontage'].fillna(train.LotFrontage.median())
```

```python
train.isna().sum()[train.isna().sum()>0]
```

```python
train["Electrical"]=train['Electrical'].fillna(train.Electrical.mode()[0])
```

```python
train["TotalSF"]=train['TotalBsmtSF']+train['1stFlrSF']+train['2ndFlrSF']
```

```python
train
```

```python
train['Age']=train["YrSold"]-train['YearBuilt']
```

```python
train['RemodAge']=train["YrSold"]-train['YearRemodAdd']
```

```python
train['HasPool'] = (train['PoolArea'] > 0) | (train['PoolQC'] != "None")
```

```python
train['HasPool']
```

```python
train['HasGarage'] = (train['GarageCars'] > 0) | (train['GarageType'] != "None")
```

```python
train['TotalBath'] = (
    train['FullBath'] +
    0.5 * train['HalfBath'] +
    train['BsmtFullBath'] +
    0.5 * train['BsmtHalfBath']
)
```

```python
train['TotalBath']
```

```python
train['IsRemodeled'] = train['YearRemodAdd'] != train['YearBuilt']
```

```python
train.MSSubClass
```

```python
train['HasFireplace'] = (train['Fireplaces'] > 0) | (train['FireplaceQu'] != "None")
```

```python
train.drop([
    'TotalBsmtSF', '1stFlrSF', '2ndFlrSF',
    'YearBuilt', 'YrSold', 'YearRemodAdd',
    'PoolArea', 'PoolQC',
    'GarageCars', 'GarageType',
    'FullBath', 'HalfBath', 'BsmtFullBath', 'BsmtHalfBath',
    'Fireplaces', 'FireplaceQu',
    'GarageYrBlt'
], axis=1, inplace=True)
```

```python
train.shape
```

```python
print(train)
```

```python
cat_cols = train.select_dtypes(include=['object']).columns
num_cols = train.select_dtypes(exclude=['object']).columns

print("Categorical columns:", len(cat_cols))
print(cat_cols)
```

```python
bool_cols = ['HasPool', 'HasGarage', 'IsRemodeled', 'HasFireplace']
train[bool_cols] = train[bool_cols].astype(int)
```

```python
cols_to_drop = [
    'TotalBsmtSF', '1stFlrSF', '2ndFlrSF',  # replaced by TotalSF
    'YearBuilt', 'YearRemodAdd', 'YrSold',  # replaced by Age, RemodAge, IsRemodeled
    'FullBath', 'HalfBath', 'BsmtFullBath', 'BsmtHalfBath',  # replaced by TotalBath
    'PoolArea', 'PoolQC',  # replaced by HasPool
    'GarageCars', 'GarageType'  # replaced by HasGarage
]
train_encoded.drop(columns=cols_to_drop, inplace=True, errors='ignore')
```

```python
train_encoded.shape
```

```python
bool_cols = train_encoded.select_dtypes('bool').columns
train_encoded[bool_cols] = train_encoded[bool_cols].astype(int)
```

```python
train_encoded.select_dtypes('bool').columns
```

```python
print(train_encoded)
```

```python
import numpy as np
import pandas as pd

# Copy the encoded dataset
train_clean = train_encoded.copy()

# Define preferred numeric columns for Z-score filtering
preferred_cols = [
    'SalePrice', 'LotArea', 'TotalSF', 'OverallQual', 'OverallCond',
    'BsmtFinSF1', '1stFlrSF', '2ndFlrSF', 'BedroomAbvGr',
    'TotalBath', 'GarageArea', 'Age', 'RemodAge'
]

# Keep only columns that actually exist in your dataset
important_cols = [col for col in preferred_cols if col in train_clean.columns]

# Compute Z-scores only for available columns
z_scores = np.abs((train_clean[important_cols] - train_clean[important_cols].mean()) / train_clean[important_cols].std())

# Remove rows where any of these columns have |Z| > 3
train_clean = train_clean[(z_scores < 3).all(axis=1)]

print("Before removing outliers:", train_encoded.shape)
print("After removing outliers:", train_clean.shape)
```

```python
train_clean['SalePrice']
```

```python
from sklearn.preprocessing import StandardScaler

# Make a copy to avoid modifying original data
train_scaled = train_clean.copy()

# Select numeric columns only
numeric_cols = train_scaled.select_dtypes(include=['float64', 'int64']).columns

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform numeric columns
train_scaled[numeric_cols] = scaler.fit_transform(train_scaled[numeric_cols])

# Show summary
print("✅ Standardization done successfully!")
print(train_scaled[numeric_cols].describe().T.head(10))  # view first 10 standardized columns
```

```python
train_scaled[numeric_cols].mean().round(2).head()

```

```python
x=train_scaled.drop('SalePrice',axis=1)
y=train_clean['SalePrice']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python

from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import make_scorer, mean_squared_error, r2_score
from xgboost import XGBRegressor
```

```python

y_train_log = np.log1p(y_train)


def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmse_scorer = make_scorer(rmse, greater_is_better=False)
r2_scorer = make_scorer(r2_score)
```

```python
base_models = {
    "Linear": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.001),
    "RandomForest": RandomForestRegressor(n_estimators=100, random_state=42),
    "XGBoost": XGBRegressor(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        tree_method='hist'
    )
}
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
results = {}
```

```python
for name, model in base_models.items():
    bagging = BaggingRegressor(
        estimator=model,
        n_estimators=10,
        max_samples=0.8,
        random_state=42
    )

    # Cross-validation for RMSE
    rmse_scores = cross_val_score(
        bagging, x_train, y_train_log,
        cv=kf,
        scoring=rmse_scorer
    )

    # Cross-validation for R² score
    r2_scores = cross_val_score(
        bagging, x_train, y_train_log,
        cv=kf,
        scoring=r2_scorer
    )

    results[name] = {
        "Mean Log RMSE": -rmse_scores.mean(),
        "Mean R² (Accuracy)": r2_scores.mean()
    }
```

```python

print(f"{'Model':15} | {'Log RMSE (Loss)':>15} | {'R² Score (Accuracy)':>20}")
print("-" * 55)

for name, vals in results.items():
    print(f"{name:15} | {vals['Mean Log RMSE']:<15.4f} | {vals['Mean R² (Accuracy)']:<20.4f}")
```
